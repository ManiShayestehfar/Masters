```python
def n_k_generator(n):
    n_ls = (n-1)*list([n])
    k_ls = list(range(1, n))
    return n_ls, k_ls

activ_func = TANH

for N in range(2,16):
    N_list, k_list = n_k_generator(N)

    tick = time.time()
    A = grid_interaction_graph(N_list, k_list, [activ_func], tol=1e-10, save=f"outputs/{activ_func.__name__}/N={N}_{activ_func.__name__}")
    tock = time.time()
    print(f"Time taken: {tock-tick:.4f} seconds")
    with open(f"outputs/{activ_func.__name__}/pickles/N={N}_{activ_func.__name__}.pkl", "wb") as f:
        pickle.dump(A, f, protocol=pickle.HIGHEST_PROTOCOL)
```
